Build a predictor funciton to predict prices of home in different cities
1) With 3400 sqr ft area in west windsor
2) 2800 sqr ft home in robbinsville

```python
import pandas as pd
```

```python
df = pd.read_csv("homeprices.csv")
df
```

```python
dummies_columns = pd.get_dummies(df.town)
dummies_columns #One Hot Encoding
#Getting Dummy Variables for Solving this types of problems
```

```python
#Now Concatinating The new frame with my old ones
merged_df = pd.concat([df,dummies_columns],axis = "columns")#Adding Old and New dataframe as axis as a column
merged_df
```

```python
final = merged_df.drop(['town','west windsor'],axis = 'columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
 X = final.drop('price',axis = 'columns')
X
```

```python
y = final.price
y
```

```python
model.fit(X,y)
```

```python
 model.predict([[2800,0,1]])
```

```python
model.predict([[3300,0,0]])
```

```python
model.score(X,y)
```
