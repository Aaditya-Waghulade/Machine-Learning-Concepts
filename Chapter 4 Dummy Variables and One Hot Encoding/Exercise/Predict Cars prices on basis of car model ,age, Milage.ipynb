```python
import pandas as pd
```

```python
carprice = pd.read_csv("carprice.csv")
carprice
```

```python
#Now making Dummy Columns with Pandas
dummies = pd.get_dummies(carprice["Car Model"])
dummies
```

1) False = 0
2) True = 1

```python
#Now Concatinating the New Dataframe of dummies with Old Dataframe
merged= pd.concat([carprice,dummies],axis='columns')
merged
```

```python
#Droping Two Columns
finale = merged.drop(['Car Model',"Mercedez Benz C class"],axis = 'columns')
finale
```

Going to define car models in numerical to get easy
1) Mercedes Benz C class = 0,0
2) Audi A5 = 1,0
3) BMW X5 = 0,1

```python
#Now Storing this values in x and y
X = finale.drop('Sell Price($)',axis = "columns") #Droped Price Column Because we are going to predict price which is dependable column
X
```

```python
y = finale['Sell Price($)'] #Storing Prices in y axis
```

```python
#Now Creating Model with Linear Regression
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[45000,4,0,0]])
```
