```python
%matplotlib inline
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits#In Sklearn we have datasets and we had loaded the handwritten modeule of load_digitts which are from 0 to 9
digits = load_digits()
```

```python
dir(digits) #this function will shows all the present element in load_digits()
```

```python
digits.images[10] #It will throw the array as output 
```

```python
#So we cannot Visible this image so we will use matplotlib for visualising this images
plt.matshow(digits.images[0 ])#This is how we use matshow to show our Array into Image
```

```python
#For Printing 5 Images then
for i in range(5):
    plt.matshow(digits.images[i])
```

```python
#Using target function
digits.target[:5]#This defines our Target total values
```

```python
from sklearn.linear_model import LogisticRegression#1)Importing Logistic Regressoin
model = LogisticRegression()
```

```python
#2) Import train_test_split 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(digits.data,digits.target,test_size=.2)
# 1) Our x = digits.data
# 2) y = digtis.target
```

```python
len(X_train)#Total Numbers of samples Training data from digits.data
```

```python
len(X_test)#Remaining samples of data for tests
```

```python
#Total Samples are
len(digits.data) 
```

```python
model.fit(X_train,y_train) #we fit our Trained sample in our LogisticRegression() function
```

```python
model.score(X_test,y_test)
```

```python
model.predict([digits.data[16]])
```
