```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
# 1) Importing All Libraries we are goind to need
```

```python
#2)Reading CSV file we have already
canada_income_file = pd.read_csv("canada_per_capita_income.csv")
#canada_income_file
```

```python
#4) Ploting the graph on an exhisting dataset we have
%matplotlib inline
plt.xlabel("Year")
plt.ylabel("per capita income (US$)")
plt.scatter(canada_income_file.year,canada_income_file.per_capita_income,color='red',marker="*")
```

```python
# 5) Writing Linear Regression formulla which will be suitable for Predicting Percapita income for year 2020 or any other year
regression = linear_model.LinearRegression()
regression.fit(canada_income_file[['per_capita_income']],canada_income_file.year)
```

```python
# 6) Now ellaborating the formula for predicting Per_capita_income is to 
# per_capita_income = coeficient * year + intercept
regression
```

```python
regression.coef_#Finding Coeficient
```

```python
regression.intercept_#finding Intercept
```

```python
regression.predict(([[2020]]))#Predecting for 2020
```
